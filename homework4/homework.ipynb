```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib
import textwrap
%matplotlib inline
#initialize the varaince and mean of likelihood and prior distribution
mu = 2
sigma = 1
data = np.random.normal(mu,sigma,500)
```

```python
#compute result
f=[]
ML = []
MAP = []
mu_prior = 2
sigma_prior = 0.5
re_prior = np.random.normal(mu_prior,sigma_prior,1)[0]
for i in range(500):
    f.append(data[i])
    sigma_post= math.sqrt(sigma**2*sigma_prior**2/(sigma**2+len(f)*sigma_prior**2))
    mu_post = sigma**2*mu_prior/(len(f)*sigma_prior**2+sigma**2)+len(f)*sigma_prior**2*(sum(f)/len(f))/(len(f)*sigma_prior**2+sigma**2)
    re_like = data[i]
    re_post = re_like*re_prior
    mu_ML = sum(f)/len(f)
    mu_MAP = mu_post
    re_prior = re_post
    #if i%10==0:
        #print(str(i)+'------------------------')
        #print('ML:'+str(mu_ML))
        #print('MAP:'+str(mu_MAP))
    ML.append(mu_ML)
    MAP.append(mu_MAP)
plt.figure()
plt.plot(ML,label='ML')
plt.legend(loc='upper right')
plt.plot(MAP,label='MAP')
plt.legend(loc='upper right')
plt.title('Likelihood variance = 1')
plt.savefig('lv=1.jpg')
plt.show()
```

```python
c=[]
for n in range(500):
    c.append(MAP[n]-ML[n])
print(len(c))
```

```python
plt.plot(c)
plt.title('Difference between ML and MAP (Incorrect prior mean)')
plt.show()
```
